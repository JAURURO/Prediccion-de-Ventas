#1-. Histograma para 'Item_Weight

```python
import seaborn as sns
import matplotlib.pyplot as plt

#Estilo
sns.set(style="darkgrid")
#Tamaño de la figura
plt.figure(figsize=(9, 7))

# Histograma para 'Item_Weight'
sns.histplot(data=ventas1, x='Item_Weight', bins=30, color='skyblue',kde=True)
plt.xlabel('Peso del producto')
plt.ylabel('Frecuencia')
plt.title('Distribución del peso del producto')
plt.show()
```


#2-.Boxplot para ver resúmenes estadísticos de diversas características en el conjunto de datos.

```python
# Boxplot para 'Item_Weight' y 'Item_MRP'
sns.boxplot(data=ventas1[['Item_Weight', 'Item_MRP','Item_Visibility']])
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.title('Boxplots para el Peso  y el Precio Máximo del Producto ')
plt.show()
```

#3-.Mapa de calor de la correlación entre las características.

```python
# Mapa de calor de correlación
correlation_matrix = ventas1.corr()
# Personalizar el mapa de calor
plt.figure(figsize=(10, 8))  # Tamaño del mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='seismic',linewidths=.8)  # Anotar los valores y usar el esquema de color 'coolwarm'
plt.title('Mapa de calor de la correlación entre las características')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para una mejor visibilidad
plt.yticks(rotation=45)  # Rotar las etiquetas del eje y para una mejor visibilidad
plt.show()  # Mostrar el mapa de calor
plt.show()
```

```python
# Calcular la matriz de correlación
correlation_matrix = ventas1.corr()

# Personalizar el mapa de calor
plt.figure(figsize=(10, 8))  # Tamaño del mapa de calor
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)  # Anotar los valores y usar el esquema de color 'coolwarm'
plt.title('Correlation Heatmap')  # Título del mapa de calor
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para una mejor visibilidad
plt.yticks(rotation=45)  # Rotar las etiquetas del eje y para una mejor visibilidad
plt.show()  # Mostrar el mapa de calor
```

```python

# Crear un gráfico de barras apiladas usando Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(x='Outlet_Establishment_Year', y='Item_Outlet_Sales', hue='Outlet_Type', data=ventas_por_anio_tipo_tienda, palette='viridis')
plt.title('Ventas Totales por Año de Establecimiento de la Tienda y Tipo de Tienda')
plt.xlabel('Año de Establecimiento de la Tienda')
plt.ylabel('Ventas Totales')
plt.legend(title='Tipo de Tienda')
plt.show()
```

```python
precio_promedio_por_tipo_producto = ventas1.groupby('Item_Type')['Item_MRP'].mean().sort_values(ascending=False)
("Precio promedio por tipo de producto:")
(precio_promedio_por_tipo_producto)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=precio_promedio_por_tipo_producto.index, y=precio_promedio_por_tipo_producto.values, palette='viridis')
plt.title('Precio Promedio por Tipo de Producto')
plt.xlabel('Tipo de Producto')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=90)
plt.show()
```

```python
# Agrupar por año de establecimiento de la tienda y tipo de producto, y calcular el precio promedio
precio_promedio_por_anio_tipo_producto = ventas1.groupby(['Outlet_Establishment_Year', 'Item_Type'])['Item_MRP'].mean().reset_index()

# Crear un gráfico de barras agrupado usando Seaborn
plt.figure(figsize=(14, 8))
sns.barplot(x='Outlet_Establishment_Year', y='Item_MRP', hue='Item_Type', data=precio_promedio_por_anio_tipo_producto, palette='viridis')
plt.title('Precio Promedio por Tipo de Producto y Año de Establecimiento de la Tienda')
plt.xlabel('Año de Establecimiento de la Tienda')
plt.ylabel('Precio Promedio')
plt.xticks(rotation=90)
plt.legend(title='Tipo de Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
